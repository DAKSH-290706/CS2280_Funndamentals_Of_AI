```python
def dfs_maze_3d(maze, start, goal):

    layers = len(maze)
    rows = len(maze[0])
    cols = len(maze[0][0])

    stack = [(start, [start])]

    # Visited set to prevent infinite loops
    visited = set()

    # Continue while there are nodes to explore
    while stack:

        # Pop last inserted element (LIFO behavior)
        (z, x, y), path = stack.pop()

        # If goal is reached, return path
        if (z, x, y) == goal:
            return path

        # Process only if not visited
        if (z, x, y) not in visited:

            # Mark current cell as visited
            visited.add((z, x, y))

            # 6 possible movement directions in 3D:
            # Up, Down, Left, Right, Forward (next layer), Backward (previous layer)
            moves = [
                (0, -1, 0),  # Up
                (0, 1, 0),   # Down
                (0, 0, -1),  # Left
                (0, 0, 1),   # Right
                (1, 0, 0),   # Move to next layer
                (-1, 0, 0)   # Move to previous layer
            ]

            for dz, dx, dy in moves:

                # Calculate neighbor position
                nz = z + dz
                nx = x + dx
                ny = y + dy

                # Check boundaries
                if (0 <= nz < layers and
                    0 <= nx < rows and
                    0 <= ny < cols and
                    maze[nz][nx][ny] != 1 and
                    (nz, nx, ny) not in visited):

                    # Add neighbor to stack with updated path
                    stack.append(((nz, nx, ny), path + [(nz, nx, ny)]))

    # No path found
    return None# 3D Maze Example (2 layers)
```

```python
maze_3d = [

    [
        ['S', 0, 1],
        [0,   2, 1],
        [1,   1, 3]
    ],

    [
        [1,   2, 0],
        [1,   1, 0],
        [0,   1, 'G']
    ]
]

start = (0, 0, 0)  # Layer 0, Row 0, Col 0
goal = (1, 2, 2)   # Layer 1, Row 2, Col 2

solution = dfs_maze_3d(maze_3d, start, goal)

if solution:
    print("Path found:")
    print(solution)
else:
    print("No path found.")
```
