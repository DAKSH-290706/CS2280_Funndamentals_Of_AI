```python
def bfs_maze(maze,start,goal):
  rows = len(maze)
  cols = len(maze[0])
  visited = set()
  visited.add(start)
  queue = deque([(start,[start])])
  while queue:
    (x,y),path = queue.popleft()
    if (x,y) == goal:
      return path
    moves = [(-1,0), (1,0), (0,-1), (0,1)]
    for dx,dy in moves:
      nx, ny = x+dx, y+dy
      if (0 <= nx < rows and
          0 <= ny < cols and
          maze[nx][ny] != 1 and
          (nx, ny) not in visited):
        visited.add((nx,ny))
        queue.append(((nx,ny), path + [(nx,ny)]))
```

```python
from collections import deque
maze = [
              ['s',0,1,0],
              [ 0 ,0,1,0],
              [ 1 ,0,0,0],
              [ 1 ,1,0,'g']
              ]
start = (0,0)
goal = (3,3)
solution = bfs_maze(maze,start,goal)
if solution:
  print("path found:")
  print(solution)
else:
  print("no path found")
```
